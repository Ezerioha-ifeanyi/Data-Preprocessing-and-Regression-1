```python
# Import the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the dataset

dataset = pd.read_csv('3.3 Salaries.csv.csv')

dataset.head()
```

```python
# Split the data into dependent and independent variable

X = dataset.iloc[:, 0].values
Y = dataset.iloc[:, -1].values
```

```python
# split the dataset into train and test set

# import the libraries
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=1/3, random_state=0)
```

```python
# Fitting the linear regression to the training set

from sklearn.linear_model import LinearRegression

# create a linear regression object
lr = LinearRegression()

# convert the training data to 2D array
X_train = X_train.reshape(-1,1)

# Fit the training dataset
lr.fit(X_train, Y_train)
```

```python
# Predicting the test set results


# Reshape the test data
X_test = X_test.reshape(-1,1)

# Make the predictions
Y_pred = lr.predict(X_test)
```

```python
# Visualize the training data

plt.scatter(X_train, Y_train)
plt.plot(X_train, lr.predict(X_train),color="red")
plt.title("Salary vs Experience (Training set results)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
 # Visualize the testing data

plt.scatter(X_test, Y_test)
plt.plot(X_train, lr.predict(X_train),color="red")
plt.title("Salary vs Experience (Test set results)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
